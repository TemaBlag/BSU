**Зачем вообще считать GC-content?**

GC-состав (GC-content) — это доля гуанина (G) и цитозина (C) в ДНК-последовательности. Этот параметр имеет несколько важных применений и значений в молекулярной биологии:

- Структурные свойства ДНК: GC-состав влияет на стабильность ДНК. Гуанин и цитозин формируют три водородные связи между собой, в отличие от двух водородных связей, образуемых аденином и тимином. Поэтому последовательности с высоким GC-содержанием обычно имеют более высокую температуру плавления (Tm), что делает их более стабильными.

- Геномная структура: GC-состав может варьироваться между разными организмами и разными участками генома. Анализ GC-содержания помогает в изучении структуры генома и функциональных особенностей. Например, участки с высоким GC-содержанием часто являются геномными областями с активной экспрессией генов.

- Генетическая кодировка и регуляция: GC-состав может влиять на регуляцию генов и на частоту мутаций. Высокое или низкое GC-содержание может быть связано с определёнными видами мутационных процессов или генетическими особенностями.

- Проектирование молекулярных инструментов: При разработке праймеров для ПЦР или других молекулярных тестов важно учитывать GC-состав, чтобы обеспечить оптимальную специфичность и эффективность реакций.

- Филогения и таксономия: Анализ GC-содержания может помочь в классификации организмов и в установлении филогенетических связей между ними.

### (?) **Что такое кодон? Что такое трансляция?**

**Кодон** — это последовательность из трёх нуклеотидов в молекуле мРНК (матричной РНК), которая кодирует одну аминокислоту в полипептидной цепи. В ДНК и мРНК есть четыре возможных нуклеотида (аденин (A), тимин (T), гуанин (G), цитозин (C) в ДНК и аденин (A), урацил (U), гуанин (G), цитозин (C) в мРНК). Поскольку кодоны состоят из трёх нуклеотидов, существует 64 возможных кодонов (4^3), из которых 61 кодон кодирует аминокислоту, а три являются стоп-кодонами, сигнализирующими завершение синтеза белка.

**Трансляция** — это процесс, в ходе которого информация, закодированная в молекуле
мРНК, используется для синтеза полипептидной цепи, которая в свою очередь формирует белок. Трансляция происходит в рибосомах и включает несколько ключевых этапов:

- Инициация

- Элонгация

- Терминация
